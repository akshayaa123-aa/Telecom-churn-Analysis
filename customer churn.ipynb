### converted 0 and 1 values of senior citizen to yes/no to make it easy to understand

```python
def conv(value):      #create functoin (convert name ka)      # if u want to know convesion of the values in the column 
    if value == 1:          # aur ye jis column ko check karna hai usi ka function banao
        return "Yes"
    else:
        return "No"
    
    
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv) # jis column ko check karna hai usi ka name likho
```

```python

# Columns list
columns = ['PhoneService', 'MultipleLines', 'InternetService', 
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV', 'StreamingMovies']        #columns

# Create a figure and subplots grid
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(11, 11))

# Flatten the axes array to easily iterate over
axes = axes.flatten()

# Loop through each column and create a countplot
for i, col in enumerate(columns):          #enumerate ka use index show karne ke liye hota hai
    sns.countplot(x=col, data=df, ax=axes[i], hue = df['Churn'])   #hue ka use graph pe color change karne ke liye hota hai
    axes[i].set_title(f"Countplot of {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Count")

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

### the majority of customer who do not churn tend to have service like phoneservice, internetservices, and online security enabled

```python
plt.figure(figsize = (6, 4))
ax = sns.countplot(x = "PaymentMethod", data = df, hue = 'Churn') # ax ka use graph pe count show karne ke liye hota hai
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])           # ax.bar_label ka use count show karne ke liye hota hai # containers[0] ka use count show karne ke liye hota hai
plt.title("Churned Customers by Payment Method")
plt.xticks(rotation = 45)
plt.show()
```
